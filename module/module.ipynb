````python
#Task 6
from textblob import TextBlob
import nltk
import urllib.request
nltk.download("punkt")

file = urllib.request.urlopen("https://www.gutenberg.org/cache/epub/15/pg15.txt")

text = ""
for line in file:
	text += line.decode("utf-8")

blob = TextBlob(text)
````

````python
#Task 7
print(f"Text: {blob.sentiment}\n")
sentences = blob.sentences[441:451]
print(f"Sentence: ")
for i, sentence in enumerate(sentences):
  print(f"{i}: {sentence.sentiment}")
````

````python
#Task 8
from textblob import Word

for sentence in sentences:
    for word in sentence.words:
      w = Word(word[:-1] + 'a')
      print(f"Wrong word: {w}, spellcheck: {w.spellcheck()}")
````
