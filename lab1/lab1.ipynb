```python
import numpy as np
import pandas as pd

np.set_printoptions(precision=2)

data = pd.read_csv("https://raw.githubusercontent.com/Stef4nio/PZNM/main/lab1/Twitch_game_data.csv", encoding = "ISO-8859-1")

display(data)
```

Міняємо всі Nan`и на нулі:

```python
data.fillna(0.0)
```

Додамо стовпчик, який буде характеризувати середній час одного стріму гри і виведемо відсортований датафрейм по доданому параметру:

```python
data["Avg_Stream_Duration"] = data["Hours_streamed"] / pd.to_numeric(data["Streamers"], errors="coerce")

display(data[["Game","Hours_streamed", "Streamers", "Avg_Stream_Duration"]].sort_values(by=["Avg_Stream_Duration"], ascending=True))
```
