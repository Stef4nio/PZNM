```python
import numpy as np
import pandas as pd
import tensorflow as tf
import pathlib
import random
import keras
import keras.models as M
import keras.layers as L
import keras.backend as K

from google.colab import files, drive
from sklearn.model_selection import train_test_split
```

```python
drive.mount('/content/drive')
```

```python
csvs =  sorted(str(p) for p in pathlib.Path('/content/drive/MyDrive/FontsPZNM/fonts').glob("*.csv"))

data = pd.DataFrame()
for csv in csvs[::2]:
  data = data.append(pd.read_csv(csv))

display(data.iloc[random.sample(range(0, len(data.index)), 10)])
```

```python
factorized = pd.factorize(data['font'])
data['font'] = factorized[0]
unique_cats_count = len(factorized[1])

print(f"Unique categories: {unique_cats_count}")

display(data.iloc[random.sample(range(0, len(data.index)), 10)])
```

```python
train, test = train_test_split(data, test_size=0.15)
```

```python
train_y = np.array(train["font"])
train_x = train.drop(columns=["font", "fontVariant"])

test_y = np.array(test["font"])
test_x = test.drop(columns=["font", "fontVariant"])
```

```python
train_y_oh = keras.utils.to_categorical(train_y, unique_cats_count)
test_y_oh = keras.utils.to_categorical(test_y, unique_cats_count)
```

```python
columns_for_learning = len(data.columns) - 2;
number_of_hidden_neurons = len(data.index) / (2 * (columns_for_learning + unique_cats_count))

print(number_of_hidden_neurons)

K.clear_session()

model = M.Sequential()
model.add(L.Normalization())
model.add(L.Dense(columns_for_learning, activation='relu'))
model.add(L.Dense(number_of_hidden_neurons, activation='elu'))
model.add(L.Dense(number_of_hidden_neurons, activation='elu'))
model.add(L.Dense(unique_cats_count, activation='softmax'))
```

```python
model.compile(loss= 'categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
results = model.fit(train_x, train_y_oh, batch_size=64, epochs=20, validation_data=(test_x, test_y_oh))
```

```python
print(f"Accuracy: {np.max(results.history['accuracy'])}")
```
